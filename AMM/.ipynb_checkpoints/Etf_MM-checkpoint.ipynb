## Squewed quotes

**skew definition** : 

$s_t = tanh( \lambda * \frac{q_t}{q^*} )$ with $\lambda$ sensitivity, $q^*$ inventory_ideal_size, $q_t$ current inventory

$Bid_t = Mid_t \times (1 - \frac{\Delta_t}{2} - s_t)$

$Bid_t = Mid_t \times (1 + \frac{\Delta_t}{2} + s_t)$

```python
def skewed_quote(current_price, base_spread, inventory,
                 inventory_ideal_size, sensitivity, max_skew):
    """
    Compute skewed quote based on inventory.
    
    Args:
        current_price (float): current mid/close price
        base_spread (float): base spread (as fraction of price, e.g. 0.02 = 2%)
        inventory (float): current inventory in $ (position_volume * price)
        inventory_ideal_size (float): $ inventory level you're comfortable with
        sensitivity (float): how aggressively you react to inventory
        max_skew (float): max skew (as fraction of price, e.g. 0.02 = 2%)
        
    Returns:
        (bid_price, offer_price)
    """
    raw_skew = sensitivity * inventory / inventory_ideal_size
    skew = current_price * max_skew * np.tanh(raw_skew)
    spread = current_price * base_spread  
    
    bid_price = current_price + skew - spread
    offer_price = current_price + skew + spread
    return bid_price, offer_price

def compute_hedge_amount(current_positions, prices, date):
    """
    compute exposure adjusted to ETF for hedging purpose
    """
    beta_dict = {"DIA": 1,'YM=F':1}  # extend later if you hedge other tickers

    exposure = 0.0
    for position in current_positions.values():
        beta = beta_dict.get(position.ticker, 1.0)  # default beta = 1
        exposure += (
            position.position_volume
            * beta
            * prices.loc[date, position.ticker]
        )

    hedge_trade_direction = "buy" if exposure <= 0 else "sell"
    return exposure, hedge_trade_direction


def compute_futures_hedge(exposure, fut_price, multiplier=5.0):
    desired_contracts = abs(exposure) / (multiplier * fut_price)
    desired_contracts_rounded = int(np.round(desired_contracts))
    #delta_contracts = desired_contracts_rounded - current_contracts
    return desired_contracts_rounded
```

```python
# create market maker & hedge fund
ExchangeSim = Exchange(df_prices)  # prices must contain the hedge ticker
mm = MarketMaker.mm()
hf = HedgeFund(seed=41)  # same interface as in the assessment

hedge_ticker = "YM=F"           
hedge_threshold = 10_000

# 16: check risk, skew prices, quote, log, show to HF, log completed trades
quoted_trades = []

for matched in request_with_prices:
    ticker = matched[0][0]
    trade_volume = matched[0][2]
    ref_price = matched[1]
    date = matched[0][1]
    current_price = matched[1]

    # --- current inventory in $ ---
    # safely get current position (0 if not present yet)
    pos = mm.current_positions.get(ticker, Position(ticker, 0.0))
    current_volume = pos.position_volume
    current_inventory = current_volume * current_price

    # --- compute skewed bid/ask ---
    bid_price, offer_price = skewed_quote(
        current_price=current_price,
        base_spread=0.002,          # 2% base spread
        inventory=current_inventory,
        inventory_ideal_size=1_000_000,
        sensitivity=1.0,
        max_skew=0.02              # up to 2% skew
    )

    # 3. Create quote
    qt = QuotedTrade(
        ticker=ticker,
        trade_volume=trade_volume,
        ref_price=ref_price,
        bid_price=bid_price,
        offer_price=offer_price,
        date=date
    )
    quoted_trades.append(qt)

    # 4. Log quote
    mm.add_quoted_trade(qt)

    # 5. Show quote to HF
    response = hf.show(qt)

    # 6. Process response -> CompletedTrade if not refused
    if response.hf_action != "refuse":
        ct = CompletedTrade(
            ticker=response.ticker,
            trade_volume=response.trade_volume,
            trade_price=response.trade_price,
            mm_action=mm_dict[response.hf_action],
            ref_price=response.ref_price,
            bid_price=response.bid_price,
            offer_price=response.offer_price,
            date=response.date
        )
        mm.add_trade(ct)

        '''----------------------------- NEW LOGIC -----------------------------'''
        exposure, hedge_trade_direction = compute_hedge_amount(
                                            mm.current_positions,
                                            df_prices,
                                            response.date
                                            )

        # Optional: only hedge if exposure significant
        if abs(exposure) > hedge_threshold:

            # price of the hedge ETF at this time
            fut_price = df_prices.loc[response.date, hedge_ticker]

            # number of units of hedge ETF to trade to offset exposure
            hedge_units = compute_futures_hedge(
                                            exposure, 
                                            fut_price, 
                                            multiplier=5.0)

            ExchangeTrade_object = ExchangeTrade(
                ticker=hedge_ticker,
                trade_volume=hedge_units,
                ref_price=fut_price,
                action=hedge_trade_direction,
                date=response.date
            )

            Executed_trade = ExchangeSim.execute(ExchangeTrade_object)

            mm.update_ETF_position(Executed_trade)
```
